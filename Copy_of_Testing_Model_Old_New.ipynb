### Air Pressure

```python
import pandas as pd
import requests

# URL API dengan parameter
url = "https://data.mertani.co.id/sensors/records"
params = {
    "sensor_company_id": "b0c04267-9c29-436f-a6e4-9b768cec4460",
    "start": sdate1,
    "end": edate1
}

# Token JWT
token = access_token
# Header dengan token JWT
headers = {"Authorization": f"Bearer {token}"}

# Kirim permintaan GET ke API dengan parameter dan header
response = requests.get(url, params=params, headers=headers)

# Initialize an empty list to store dictionaries
airpressrnd_data = []

# Periksa status kode respons
if response.status_code == 200:
    data = response.json()  # Jika respons adalah JSON
    # Lakukan sesuatu dengan data
    # print(data)

    # Extract and print the "date," "time," and "value_calibration" fields from the records
    records = data["data"]["data"][0]["sensor_records"]
    for record in records:
        date = record["date"]
        datetime = record["datetime"]
        value_calibration = record["value_calibration"]
       # print(f"Date: {date}, Time: {time}, Value Calibration: {value_calibration}")

        # Append data as dictionary to airhum_data list
        airpressrnd_data.append({"Date": date, "Datetime": datetime, "airpressrnd": value_calibration})

airpressrnd = pd.DataFrame(airpressrnd_data)
print(airpressrnd)
```

### Daily

```python
# Menghitung nilai maksimum harian
max_humrnd = airhumrnd.groupby('Date')['airhumrnd'].max().reset_index()
max_humrnd = max_humrnd.rename(columns={'airhumrnd': 'max_airhum'})

# Menghitung nilai minimum harian
min_humrnd = airhumrnd.groupby('Date')['airhumrnd'].min().reset_index()
min_humrnd = min_humrnd.rename(columns={'airhumrnd': 'min_airhum'})

# Menghitung nilai rata - rata harian
avg_humrnd= airhumrnd.groupby('Date')['airhumrnd'].mean().reset_index()
avg_humrnd = avg_humrnd.rename(columns={'airhumrnd': 'avg_airhum'})

# Menghitung nilai rata - rata harian
avg_wsrnd= wsrnd.groupby('Date')['wsrnd'].mean().reset_index()
avg_wsrnd = avg_wsrnd.rename(columns={'wsrnd': 'avg_ws'})

# Menghitung nilai maksimum harian
max_temprnd = airtemprnd.groupby('Date')['airtemprnd'].max().reset_index()
max_temprnd = max_temprnd.rename(columns={'airtemprnd': 'max_airtemp'})

# Menghitung nilai minimum harian
min_temprnd = airtemprnd.groupby('Date')['airtemprnd'].min().reset_index()
min_temprnd = min_temprnd.rename(columns={'airtemprnd': 'min_airtemp'})

# Menghitung nilai rata - rata harian
avg_temprnd= airtemprnd.groupby('Date')['airtemprnd'].mean().reset_index()
avg_temprnd = avg_temprnd.rename(columns={'airtemprnd': 'avg_airtemp'})

# Menghitung nilai maksimum harian
max_pressrnd = airpressrnd.groupby('Date')['airpressrnd'].max().reset_index()
max_pressrnd = max_pressrnd.rename(columns={'airpressrnd': 'max_airpress'})

# Menghitung nilai minimum harian
min_pressrnd = airpressrnd.groupby('Date')['airpressrnd'].min().reset_index()
min_pressrnd = min_pressrnd.rename(columns={'airpressrnd': 'min_airpress'})

# Menghitung nilai rata - rata harian
avg_pressrnd= airpressrnd.groupby('Date')['airpressrnd'].mean().reset_index()
avg_pressrnd = avg_pressrnd.rename(columns={'airpressrnd': 'avg_airpress'})
```

```python
# Merge variabel microclimate
iklimrnd=pd.merge(avg_humrnd, max_humrnd,on='Date',how='outer')
iklimrnd=pd.merge(iklimrnd, min_humrnd,on='Date',how='outer')
iklimrnd=pd.merge(iklimrnd, avg_temprnd,on='Date',how='outer')
iklimrnd=pd.merge(iklimrnd, max_temprnd,on='Date',how='outer')
iklimrnd=pd.merge(iklimrnd, min_temprnd,on='Date',how='outer')
iklimrnd=pd.merge(iklimrnd, avg_wsrnd,on='Date',how='outer')
iklimrnd=pd.merge(iklimrnd, avg_pressrnd,on='Date',how='outer')
iklimrnd=pd.merge(iklimrnd, max_pressrnd,on='Date',how='outer')
iklimrnd=pd.merge(iklimrnd, min_pressrnd,on='Date',how='outer')
iklimrnd
```
